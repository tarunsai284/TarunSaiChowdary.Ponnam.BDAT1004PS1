## Question 2
#### a) How many letters are there in 'Supercalifragilisticexpialidocious'?
#### b. Does 'Supercalifragilisticexpialidocious' contain 'ice' as a substring?
#### c. Which of the following words is the longest: Supercalifragilisticexpialidocious, Honorificabilitudinitatibus,  or Bababadalgharaghtakamminarronnkonn?  
#### d. Which composer comes first in the dictionary: 'Berlioz', 'Borodin', 'Brian',  'Bartok', 'Bellini', 'Buxtehude', 'Bernstein'. Which one comes last?

```python
using System;
using System.Collections.Generic;

namespace DemoApp1
{
    class Program
    {
        
        public void GetStringLength(string str)
        {
            Console.WriteLine($"No of letters in \'{str}\': {str.Length}");
        }

        public void CheckSubstring(string str, string sub)
        {
            Console.WriteLine($" Is \'{sub}\' present in the \'{str}\': {str.Contains(sub)}");
        }

        public void Longest(params string[] list)
        {
            var longest=new List<string>();
            int maxLength = 0;
            foreach(string str in list)
            {
                if (maxLength < str.Length)
                {
                    longest.Clear();
                    longest.Add(str);
                    maxLength = str.Length;
                }
                else if (maxLength == str.Length) longest.Add(str);
            }
            if (longest.Count > 1)
            {
                Console.WriteLine("Longest strings are:");
                longest.ForEach(Console.WriteLine);
                Console.WriteLine();
            }
            else Console.WriteLine($"Longest string is {longest}");
        }

        public void FistAndLast(List<string> list)
        {
            list.Sort();
            Console.WriteLine($"The first item is \"{list[0]}\" and the last item \"{list[list.Count - 1]}\" in the list of lexicographic order");
        }

        public static void Main(string[] args)
        {

            Program assignment = new Program();

            // a) Function to prints the number of letters
            assignment.GetStringLength("Supercalifragilisticexpialidocious");

            // b) Function to ice substring
            assignment.CheckSubstring("Supercalifragilisticexpialidocious", "ice");

            // c) Longest string
            assignment.Longest("Supercalifragilisticexpialidocious", "Honorificabilitudinitatibus", "Bababadalgharaghtakamminarronnkonn");

            // d) Fist and last in a dictonary
            List<String> list = new List<string>() { "Berlioz", "Borodin", "Brian", "Bartok", "Bellini", "Buxtehude", "Bernstein" };
            assignment.FistAndLast(list);
        }
    }
}
```

#### Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. By Heron's formula, the area of a triangle with side lengths a, b, and c is s(s- a)(s- b)(s- c), where s= (a+ b+ c)/2.


```python
using System;
using System.Collections.Generic;

namespace DemoApp1
{
    class Program
    {
        public void triangleArea(int a, int b, int c)
        {
            int s = (a + b + c) / 2;
            int area2 = s * (s - a) * (s - b) * (s - c);
            Console.WriteLine(Math.Sqrt(area2));
        }

        public static void Main(string[] args)
        {
            Program program = new Program();
            program.triangleArea(2, 2, 2);
        }
        
    }
}
```

#### Write a function inside(x,y,x1,y1,x2,y2) that returns True or False depending on whether the point (x,y) lies in the rectangle with lower left corner (x1,y1) and upper right corner (x2,y2)


```python
using System;

namespace DemoApp1
{
    class Program
    {
        public bool Inside(double x, double y, double x1, double y1, double x2, double y2)
        {
            bool inside = false;
            if (x > x1 && y > y1 && x < x2 && y < y2) inside = true;
            return inside;
        }

        public static void Main(string[] args)
        {
            Program program = new Program();
            // Answer a)
            Console.WriteLine(program.Inside(1, 1, 0, 0, 2, 3));
            Console.WriteLine(program.Inside(-1, -1, 0, 0, 2, 3));
            // Answer b)
            Console.WriteLine(program.Inside(1, 1, 0.3, 0.5, 1.1, 0.7) && program.Inside(1, 1, 0.5, 0.2, 1.1, 2));
        }
    }
}
```

## Question 7 Python 
#### File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic. Write a function bldcount() that reads the file with name name and reports (i.e., prints) how many patients there are in each bloodtype.

```python
def bldcount(path):
    bloodGroups = {
        "A" : 0,
        "B" : 0,
        "AB" : 0,
        "O" : 0,
        "OO" : 0
    }
    file = open(path)
    for line in file:
        for group in line.split():
            if group in bloodGroups: bloodGroups[group] += 1
    
    file.close()
    
    for group in bloodGroups:
        if bloodGroups[group]==1: 
            print ("There is one patient of blood type {}.".format(group))
        elif bloodGroups[group]==0:
            print ("There are no patients of blood type {}.".format(group))
        else:
            print ("There are {} patients of blood type {}.".format(bloodGroups[group],group))
        

bldcount("bloodtype1.txt")
```

#### Write a function curconv() that takes as input;

```python
def curconv(currency,amount):
    file = open("currencies.txt")
    exchangeRates = {}
    for line in file:
        rate = line.split()
        if len(rate) > 0:
            exchangeRates[rate[0]] = float(rate[1])
    file.close()
            
    dollarAmount = amount*exchangeRates[currency]
    print(dollarAmount)
    
curconv('EUR', 100)
curconv('JPY', 100)
        
```

#### Encryption is the process of hiding the meaning of a text by substituting letters in the message with other letters, according to some system. If the process is successful, no one but the intended recipient can understand the encrypted message. Cryptanalysis refers to attempts to undo the encryption, even if some details of the encryption are unknown (for example, if an encrypted message has been intercepted). The first step of cryptanalysis is often to build up a table of letter frequencies in the encrypted text. Assume that the string letters is already defined as 'abcdefghijklmnopqrstuvwxyz'. Write a function called frequencies() that takes a string as its only parameter, and returns a list of integers, showing the number of times each character appears in the text. Your function may ignore any characters that are not in letters.

```python
def frequencies(target):
    frequencyList = [0]*26
    for char in target:
        asciiValue = ord(char.lower())-97
        if asciiValue >= 0 and asciiValue <= 25: 
            frequencyList[asciiValue] += 1
    
    print(frequencyList)
    
frequencies("The quick red fox got bored and went home.")
frequencies("apple")
```
